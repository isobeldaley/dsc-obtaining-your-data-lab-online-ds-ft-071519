Start by investigating the database stored in lego.db and joining the tables into a unified dataset!

> **Hint:** use this sql statement to preview the tables in an unknown database:
```sql
SELECT name FROM sqlite_master
             WHERE type='table'
             ORDER BY name;
```

````python
# Your code here
import sqlite3
import pandas as pd

conn = sqlite3.connect('lego.db')
c = conn.cursor()

````

````python
# Preview tables within the database
c.execute("""SELECT name FROM sqlite_master
                WHERE type='table'
                ORDER BY name;""")
df = pd.DataFrame(c.fetchall())
df
````

````python
# Preview each table to confirm which key tables should be joined on
c.execute("""SELECT *
                FROM product_reviews""")
df = pd.DataFrame(c.fetchall())
df.columns = [x[0] for x in c.description]
df.head()
````

````python
# Join all tables in a dataset

c.execute("""SELECT *
                FROM product_details
                JOIN product_info
                USING(prod_id)
                JOIN product_pricing
                USING(prod_id)
                JOIN product_reviews
                USING(prod_id)""")

df = pd.DataFrame(c.fetchall())
df.columns = [x[0] for x in c.description]
df.head()
````
